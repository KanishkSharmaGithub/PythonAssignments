```python
4. Create an Element class with the instance attributes name, symbol, and number. Create a class
object with the values 'Hydrogen', 'H', and 1.


# Define the Element class
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number

# Create a class object with the values 'Hydrogen,' 'H,' and 1
hydrogen = Element('Hydrogen', 'H', 1)

# Access and print the attributes of the hydrogen object
print("Name:", hydrogen.name)
print("Symbol:", hydrogen.symbol)
print("Number:", hydrogen.number)


We define the Element class with an __init__ method that initializes the object with the name, symbol, and number instance attributes.

We create a class object named hydrogen by calling the Element class constructor with the values 'Hydrogen,' 'H,' and 1.

We access and print the attributes of the hydrogen object, which will display the values assigned during object creation.

When you run this code, it will print the name, symbol, and number of the hydrogen element object:
    
Name: Hydrogen
Symbol: H
Number: 1
```

```python
5. Make a dictionary with these keys and values: 'name': 'Hydrogen', 'symbol': 'H, 'number': 1. Then,
create an object called hydrogen from class Element using this dictionary.


# Define the dictionary with keys and values
element_info = {'name': 'Hydrogen', 'symbol': 'H', 'number': 1}

# Create an object 'hydrogen' from class 'Element' using the dictionary
hydrogen = Element(**element_info)

# Access and print the attributes of the 'hydrogen' object
print("Name:", hydrogen.name)
print("Symbol:", hydrogen.symbol)
print("Number:", hydrogen.number)

We define a dictionary named element_info with keys 'name', 'symbol', and 'number', and their respective values.

We create an object hydrogen from the Element class using the dictionary unpacking syntax **element_info. This passes the dictionary values as keyword arguments to the class constructor.

We access and print the attributes of the hydrogen object, which will display the values from the dictionary.
```

```python
6. For the Element class, define a method called dump() that prints the values of the objectâ€™s
attributes (name, symbol, and number). Create the hydrogen object from this new definition and
use dump() to print its attributes.


# Define the Element class with the dump() method
class Element:
    def __init__(self, name, symbol, number):
        self.name = name
        self.symbol = symbol
        self.number = number

    def dump(self):
        print("Name:", self.name)
        print("Symbol:", self.symbol)
        print("Number:", self.number)

# Create the hydrogen object from the updated Element class
hydrogen = Element('Hydrogen', 'H', 1)

# Use the dump() method to print the attributes of the hydrogen object
hydrogen.dump()
```

```python
8. Modify Element to make the attributes name, symbol, and number private. Define a getter
property for each to return its value.


# Define the Element class with private attributes and getter properties
class Element:
    def __init__(self, name, symbol, number):
        self._name = name  # Private attribute with a leading underscore
        self._symbol = symbol  # Private attribute with a leading underscore
        self._number = number  # Private attribute with a leading underscore

    # Getter property for 'name'
    @property
    def name(self):
        return self._name

    # Getter property for 'symbol'
    @property
    def symbol(self):
        return self._symbol

    # Getter property for 'number'
    @property
    def number(self):
        return self._number

# Create the hydrogen object from the updated Element class
hydrogen = Element('Hydrogen', 'H', 1)

# Access and print the attributes using the getter properties
print("Name:", hydrogen.name)
print("Symbol:", hydrogen.symbol)
print("Number:", hydrogen.number)
```

```python
9. Define three classes: Bear, Rabbit, and Octothorpe. For each, define only one method: eats(). This
should return 'berries' (Bear), 'clover' (Rabbit), or 'campers' (Octothorpe). Create one object from
each and print what it eats.


# Define the Bear class with an eats() method
class Bear:
    def eats(self):
        return 'berries'

# Define the Rabbit class with an eats() method
class Rabbit:
    def eats(self):
        return 'clover'

# Define the Octothorpe class with an eats() method
class Octothorpe:
    def eats(self):
        return 'campers'

# Create objects from each class
bear = Bear()
rabbit = Rabbit()
octothorpe = Octothorpe()

# Print what each animal eats
print("Bear eats:", bear.eats())
print("Rabbit eats:", rabbit.eats())
print("Octothorpe eats:", octothorpe.eats())

```

```python
10. Define these classes: Laser, Claw, and SmartPhone. Each has only one method: does(). This
returns 'disintegrate' (Laser), 'crush' (Claw), or 'ring' (SmartPhone). Then, define the class Robot that
has one instance (object) of each of these. Define a does() method for the Robot that prints what its
component objects do.


# Define the Laser class with a does() method
class Laser:
    def does(self):
        return 'disintegrate'

# Define the Claw class with a does() method
class Claw:
    def does(self):
        return 'crush'

# Define the SmartPhone class with a does() method
class SmartPhone:
    def does(self):
        return 'ring'

# Define the Robot class with components Laser, Claw, and SmartPhone
class Robot:
    def __init__(self):
        self.laser = Laser()
        self.claw = Claw()
        self.smartphone = SmartPhone()

    # Define a does() method for the Robot that prints what its components do
    def does(self):
        print(f"Laser does: {self.laser.does()}")
        print(f"Claw does: {self.claw.does()}")
        print(f"SmartPhone does: {self.smartphone.does()}")

# Create a Robot object
robot = Robot()

# Call the does() method of the Robot to print what its components do
robot.does()
```
