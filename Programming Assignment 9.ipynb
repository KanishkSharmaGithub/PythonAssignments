```python
1. Write a Python program to check if the given number is a Disarium Number?


# Function to check if a number is a Disarium number
def is_disarium_number(num):
    num_str = str(num)
    length = len(num_str)
    total = sum(int(digit) ** (i + 1) for i, digit in enumerate(num_str))

    return num == total

# Input from the user
try:
    num = int(input("Enter a number: "))
    
    if is_disarium_number(num):
        print(f"{num} is a Disarium number.")
    else:
        print(f"{num} is not a Disarium number.")
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

```python
3. Write a Python program to check if the given number is Happy Number?


# Function to check if a number is a happy number
def is_happy_number(num):
    seen = set()
    while num != 1 and num not in seen:
        seen.add(num)
        num = sum(int(digit) ** 2 for digit in str(num))
    return num == 1

# Input from the user
try:
    num = int(input("Enter a number: "))
    
    if is_happy_number(num):
        print(f"{num} is a happy number.")
    else:
        print(f"{num} is not a happy number.")
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

```python
5. Write a Python program to determine whether the given number is a Harshad Number?



# Function to check if a number is a Harshad Number
def is_harshad_number(num):
    digit_sum = sum(int(digit) for digit in str(num))
    return num % digit_sum == 0

# Input from the user
try:
    num = int(input("Enter a number: "))
    
    if is_harshad_number(num):
        print(f"{num} is a Harshad Number.")
    else:
        print(f"{num} is not a Harshad Number.")
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

```python
6. Write a Python program to print all pronic numbers between 1 and 100?


# Function to check if a number is a pronic number
def is_pronic_number(num):
    for n in range(int(num**0.5) + 1):
        if n * (n + 1) == num:
            return True
    return False

# Find and print pronic numbers between 1 and 100
pronic_numbers = [num for num in range(1, 101) if is_pronic_number(num)]

print("Pronic numbers between 1 and 100:")
print(pronic_numbers)
```
