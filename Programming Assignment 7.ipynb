```python
1. Write a Python Program to find sum of array?

# Function to calculate the sum of elements in an array
def find_sum(arr):
    return sum(arr)

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Calculate and print the sum of the array
    total_sum = find_sum(numbers)
    print(f"The sum of the elements in the array is: {total_sum}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
2. Write a Python Program to find largest element in an array?


# Function to find the largest element in an array
def find_largest_element(arr):
    if not arr:
        return "The array is empty."
    else:
        largest = arr[0]
        for element in arr:
            if element > largest:
                largest = element
        return largest

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Find and print the largest element in the array
    largest_element = find_largest_element(numbers)
    if isinstance(largest_element, str):
        print(largest_element)
    else:
        print(f"The largest element in the array is: {largest_element}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
3. Write a Python Program for array rotation?



# Function to rotate an array to the right by a specified number of positions
def rotate_array_right(arr, positions):
    n = len(arr)
    positions %= n  # Handle cases where positions are greater than the array length
    return arr[-positions:] + arr[:-positions]

# Input from the user
try:
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))
    positions = int(input("Enter the number of positions to rotate to the right: "))

    # Rotate the array to the right and print the result
    rotated_array = rotate_array_right(numbers, positions)
    print("Rotated array to the right:", rotated_array)
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces and a number of positions as an integer.")
```

```python
4. Write a Python Program to Split the array and add the first part to the end?


# Function to split the array and add the first part to the end
def split_and_add(arr, split_position):
    if split_position < 0 or split_position >= len(arr):
        return "Invalid split position."
    return arr[split_position:] + arr[:split_position]

# Input from the user
try:
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))
    split_position = int(input("Enter the split position (0 to n-1): "))

    # Split the array and add the first part to the end
    result = split_and_add(numbers, split_position)
    if isinstance(result, str):
        print(result)
    else:
        print("Split and added array:", result)
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces and a valid split position.")
```

```python
5. Write a Python Program to check if given array is Monotonic?


# Function to check if the array is monotonic
def is_monotonic(arr):
    increasing = decreasing = True

    for i in range(1, len(arr)):
        if arr[i] > arr[i - 1]:
            decreasing = False
        elif arr[i] < arr[i - 1]:
            increasing = False

    return increasing or decreasing

# Input from the user
try:
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Check if the array is monotonic and print the result
    if is_monotonic(numbers):
        print("The array is monotonic.")
    else:
        print("The array is not monotonic.")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```
