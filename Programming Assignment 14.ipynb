```python
Define a class with a generator which can iterate the numbers, which are divisible by
7, between a given range 0 and n.


class DivisibleBySeven:
    def __init__(self, n):
        self.n = n

    def divisible_by_seven(self):
        for i in range(self.n + 1):
            if i % 7 == 0:
                yield i

# Taking input from the user for the upper limit 'n'
n = int(input("Enter a value for 'n': "))

# Creating an instance of the class and using the generator
div_by_seven = DivisibleBySeven(n)

print("Numbers divisible by 7 between 0 and", n, ":")
for number in div_by_seven.divisible_by_seven():
    print(number)
```

```python
Define a class Person and its two child classes: Male and Female. All classes have a
method &quot;getGender&quot; which can print &quot;Male&quot; for Male class and &quot;Female&quot; for Female
class.


class Person:
    def getGender(self):
        return "Unknown"

class Male(Person):
    def getGender(self):
        return "Male"

class Female(Person):
    def getGender(self):
        return "Female"

# Creating instances and calling the getGender method
person = Person()
male = Male()
female = Female()

# Print the gender using getGender method
print("Person:", person.getGender())
print("Male:", male.getGender())
print("Female:", female.getGender())
```

```python
Please write a program to generate all sentences where subject is in [&quot;I&quot;, &quot;You&quot;] and
verb is in [&quot;Play&quot;, &quot;Love&quot;] and the object is in [&quot;Hockey&quot;,&quot;Football&quot;].


subjects = ["I", "You"]
verbs = ["Play", "Love"]
objects = ["Hockey", "Football"]

# Generate sentences using nested loops
sentences = [f"{subject} {verb} {object}" for subject in subjects for verb in verbs for object in objects]

# Display the generated sentences
for sentence in sentences:
    print(sentence)
```

```python
Please write a program to compress and decompress the string &quot;hello world!hello
world!hello world!hello world!&quot;.

To Compress the string:

import zlib

# Original string
original_string = "hello world!hello world!hello world!hello world!"

# Compressing the string
compressed_string = zlib.compress(original_string.encode())

# Printing the compressed string
print("Compressed String:", compressed_string)


To Decompress the string:
    
import zlib

# Compressed string
compressed_string = b'x\x9cKLJNJV/V\xca\xcfS\x08\x00\xbd\r\x19'

# Decompressing the string
decompressed_string = zlib.decompress(compressed_string)

# Printing the decompressed string
print("Decompressed String:", decompressed_string.decode())
```

```python
Please write a binary search function which searches an item in a sorted list. The
function should return the index of element to be searched in the list.



def binary_search(arr, x):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        # Check if x is present at mid
        if arr[mid] == x:
            return mid

        # If x is greater, ignore left half
        elif arr[mid] < x:
            left = mid + 1

        # If x is smaller, ignore right half
        else:
            right = mid - 1

    # If the element is not present in the list
    return -1

# Example usage:
sorted_list = [2, 4, 6, 8, 10, 12, 14, 16]
search_item = 10

# Perform binary search and print the result
result = binary_search(sorted_list, search_item)
if result != -1:
    print(f"Element found at index {result}.")
else:
    print("Element is not present in the list.")
```
