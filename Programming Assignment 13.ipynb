```python
Write a program that calculates and prints the value according to the given formula:
Q = Square root of [(2 * C * D)/H]
Following are the fixed values of C and H:
C is 50. H is 30.
D is the variable whose values should be input to your program in a comma-separated
sequence.



import math

# Fixed values
C = 50
H = 30

# Function to calculate Q for a single value of D
def calculate_Q(D):
    Q = math.sqrt((2 * C * D) / H)
    return Q

# Input from the user: comma-separated values of D
input_str = input("Enter values of D (comma-separated): ")
D_values = input_str.split(',')

# Calculate and print Q for each value of D
result = []
for D in D_values:
    try:
        D = float(D)
        Q = calculate_Q(D)
        result.append(Q)
    except ValueError:
        result.append("Invalid input")

print("Results for D values:", result) 
```

```python
Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
element value in the i-th row and j-th column of the array should be i*j.
Note: i=0,1.., X-1; j=0,1,Â¡Y-1.
    
    
# Function to generate a 2-dimensional array
def generate_2d_array(X, Y):
    result = []
    for i in range(X):
        row = []
        for j in range(Y):
            row.append(i * j)
        result.append(row)
    return result

# Input from the user
X = int(input("Enter the number of rows (X): "))
Y = int(input("Enter the number of columns (Y): "))

# Generate and print the 2-dimensional array
two_dimensional_array = generate_2d_array(X, Y)
for row in two_dimensional_array:
    print(row)
```

```python
Write a program that accepts a comma separated sequence of words as input and prints the
words in a comma-separated sequence after sorting them alphabetically.
Suppose the following input is supplied to the program:
without,hello,bag,world
Then, the output should be:
bag,hello,without,world



# Input from the user
input_str = input("Enter a comma-separated sequence of words: ")

# Split the input string into words, sort them, and join them back into a string
words = input_str.split(',')
sorted_words = sorted(words)
output_str = ','.join(sorted_words)

# Print the sorted words
print("Sorted words:", output_str)
```

```python
Write a program that accepts a sequence of whitespace separated words as input and prints
the words after removing all duplicate words and sorting them alphanumerically.
Suppose the following input is supplied to the program:
hello world and practice makes perfect and hello world again
Then, the output should be:
again and hello makes perfect practice world



# Input from the user
input_str = input("Enter a sequence of whitespace-separated words: ")

# Split the input string into words, remove duplicates, sort them, and join them back into a string
words = input_str.split()
unique_words = list(set(words))  # Removes duplicates
sorted_words = sorted(unique_words)
output_str = ' '.join(sorted_words)

# Print the sorted and deduplicated words
print("Sorted and deduplicated words:", output_str)
```

```python
Write a program that accepts a sentence and calculate the number of letters and digits.
Suppose the following input is supplied to the program:
hello world! 123
Then, the output should be:
LETTERS 10



# Input from the user
input_str = input("Enter a sentence: ")

# Initialize counters for letters and digits
letter_count = 0
digit_count = 0

# Iterate through each character in the input string
for char in input_str:
    if char.isalpha():
        letter_count += 1
    elif char.isdigit():
        digit_count += 1

# Print the counts of letters and digits
print("LETTERS", letter_count)
print("DIGITS", digit_count)
```
