```python
4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with
these fields: title (text), author (text), and year (integer).
    
    
import sqlite3

# Connect to the SQLite database (or create it if it doesn't exist)
connection = sqlite3.connect('books.db')

# Create a cursor object to execute SQL commands
cursor = connection.cursor()

# Define the SQL command to create the 'books' table
create_table_sql = '''
CREATE TABLE IF NOT EXISTS books (
    title TEXT,
    author TEXT,
    year INTEGER
)
'''

# Execute the SQL command to create the table
cursor.execute(create_table_sql)

# Commit the changes and close the database connection
connection.commit()
connection.close()
```

```python
3. Create a CSV file called books.csv by using these lines:
title,author,year
The Weirdstone of Brisingamen,Alan Garner,1960
Perdido Street Station,China Miéville,2000
Thud!,Terry Pratchett,2005
The Spellman Files,Lisa Lutz,2007
Small Gods,Terry Pratchett,1992



import csv

# Define the data
data = [
    ["title", "author", "year"],
    ["The Weirdstone of Brisingamen", "Alan Garner", 1960],
    ["Perdido Street Station", "China Miéville", 2000],
    ["Thud!", "Terry Pratchett", 2005],
    ["The Spellman Files", "Lisa Lutz", 2007],
    ["Small Gods", "Terry Pratchett", 1992]
]

# Create the CSV file 'books.csv' and write the data
with open('books.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("books.csv has been created.")
```

```python
5. Read books.csv and insert its data into the book table.


import sqlite3
import csv

# Connect to the SQLite database 'books.db'
connection = sqlite3.connect('books.db')

# Create a cursor object to execute SQL commands
cursor = connection.cursor()

# Open and read the 'books.csv' file
with open('books.csv', mode='r', newline='') as file:
    csv_reader = csv.reader(file)
    
    # Skip the header row (first row)
    next(csv_reader)
    
    # Iterate over the rows in the CSV file and insert them into the 'books' table
    for row in csv_reader:
        title, author, year = row
        cursor.execute("INSERT INTO books (title, author, year) VALUES (?, ?, ?)", (title, author, int(year)))

# Commit the changes and close the database connection
connection.commit()
connection.close()

print("Data from books.csv has been inserted into the 'books' table.")
```

```python
6. Select and print the title column from the book table in alphabetical order.


import sqlite3

# Connect to the SQLite database 'books.db'
connection = sqlite3.connect('books.db')

# Create a cursor object to execute SQL commands
cursor = connection.cursor()

# Execute an SQL query to select the 'title' column from the 'books' table in alphabetical order
cursor.execute("SELECT title FROM books ORDER BY title ASC")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Print the 'title' column values
for row in rows:
    print(row[0])

# Close the database connection
connection.close()
```

```python
7. From the book table, select and print all columns in the order of publication.


import sqlite3

# Connect to the SQLite database 'books.db'
connection = sqlite3.connect('books.db')

# Create a cursor object to execute SQL commands
cursor = connection.cursor()

# Execute an SQL query to select all columns from the 'books' table in the order of publication
cursor.execute("SELECT * FROM books ORDER BY year ASC")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Print the columns (title, author, and year) for each row
for row in rows:
    title, author, year = row
    print(f"Title: {title}, Author: {author}, Year: {year}")

# Close the database connection
connection.close()
```

```python
10. Increment the count field of test and print it.


from sqlalchemy import create_engine, MetaData, Table, Column, Integer, select, update

# Create a SQLAlchemy engine to connect to the database
engine = create_engine('sqlite:///books.db')

# Define the table and metadata
metadata = MetaData()
test = Table('test', metadata, autoload=True, autoload_with=engine)

# Increment the count field for a specific record
record_id = 1  # Replace with the appropriate record ID
increment_query = update(test).where(test.c.id == record_id).values(count=test.c.count + 1)

# Execute the update query
with engine.connect() as connection:
    connection.execute(increment_query)

# Retrieve and print the updated count value
select_query = select([test.c.count]).where(test.c.id == record_id)
with engine.connect() as connection:
    result = connection.execute(select_query)
    updated_count = result.scalar()

print(f"Updated count: {updated_count}")

# Close the engine (optional)
engine.dispose()
```
