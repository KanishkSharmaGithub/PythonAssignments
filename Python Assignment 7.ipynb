```python
4. From a Match item, how do you get the actual strings that match the pattern?


To get the actual strings that match the pattern from a Match object in Python (using the re module), you can use the .group()
method. The .group() method allows you to retrieve the matched substring(s) from the text.

import re

pattern = r'apple'
text = 'I have an apple and a banana.'

match = re.search(pattern, text)

if match:
    matched_string = match.group()
    print('Match found:', matched_string)
else:
    print('No match found')
```

```python
7. The findall() method returns a string list or a list of string tuples. What causes it to return one of
the two options?


The findall() method in Python's re module returns either a list of strings or a list of tuples depending on how you define
capturing groups in your regular expression pattern.

List of Strings (Non-Capturing Groups): If your regular expression pattern does not contain any capturing groups
    (i.e., no parentheses used for grouping with the (...) syntax), findall() will return a list of strings. 
Each element in the list will represent a complete match of the pattern.

List of Tuples (Capturing Groups): If your regular expression pattern contains one or more capturing groups, findall() will
    return a list of tuples. Each tuple corresponds to a complete match of the pattern, and the elements of the tuple represent
the captured groups within that match.
```

```python
10.In regular expressions, what is the difference between the + and * characters?


+ (Plus Quantifier):

The + character specifies that the preceding element must appear at least once but can also appear multiple times (one or more times).
In other words, it requires that the preceding element occurs at least once and allows for additional occurrences.

* (Asterisk Quantifier):

The * character specifies that the preceding element can appear zero or more times.
It allows for the preceding element to be optional; it may or may not appear in the matched text.
```

```python
11. What is the difference between {4} and {4,5} in regular expression?

{4}:

{4} specifies that the preceding element must appear exactly 4 times.
Example:

Pattern: a{4}
Matches: "aaaa"
Does not match: "aaa" (less than 4 "a" characters) or "aaaaa" (more than 4 "a" characters).
{4,5}:

{4,5} specifies a range for the number of repetitions of the preceding element. It means the element must appear at least 4 times but no more than 5 times.
Example:

Pattern: a{4,5}
Matches: "aaaa", "aaaaa"
```

```python
12. What do you mean by the \d, \w, and \s shorthand character classes signify in regular
expressions?


\d:

\d represents any digit from 0 to 9. It is a shorthand for the character class [0-9].
Example:

Pattern: \d\d\d
Matches: "123", "456", "789"
Does not match: "abc" (contains non-digit characters)
\w:

\w represents any word character, which typically includes alphanumeric characters (letters and digits) as well as underscores. It is a shorthand for the character class [a-zA-Z0-9_].
Example:

Pattern: \w+
Matches: "word", "123", "my_var", "A_B_C"
Does not match: "!" (contains a non-word character)
\s:

\s represents any whitespace character, including spaces, tabs, and newline characters. It is a shorthand for the character class [\t\n\f\r\p{Z}], where \p{Z} represents various Unicode space separators.
Example:

Pattern: \s+
Matches: " ", "\t", "\n", "\r\n", etc.
Does not match: "word" (contains non-whitespace characters)
These shorthand character classes make it more convenient to match common character types in regular expressions. They can be used alone or in combination with other regular expression elements to create more complex patterns for text matching and manipulation.

```

```python
13. What do means by \D, \W, and \S shorthand character classes signify in regular expressions?



\D:

\D represents any character that is NOT a digit (0-9). It is the negation of \d.
Example:

Pattern: \D+
Matches: "abcXYZ", "!@#", "word and spaces"
Does not match: "12345" (contains only digits)
\W:

\W represents any character that is NOT a word character. It is the negation of \w. It matches characters that are not letters, digits, or underscores.
Example:

Pattern: \W+
Matches: "!@#", "spaces and punctuation", "123-456"
Does not match: "word_variable" (contains word characters)
\S:

\S represents any character that is NOT whitespace. It is the negation of \s. It matches characters that are not spaces, tabs, or newline characters.
Example:

Pattern: \S+
Matches: "word", "123-456", "!@#", "charactersWithoutSpaces"
Does not match: " " (contains only whitespace)
These negated shorthand character classes are useful when you want to match characters that do not belong to a particular character class. They are often used in regular expressions to identify and extract non-standard or non-alphanumeric characters from a text or to split text based on delimiters and separators.
```

```python
14. What is the difference between .*? and .*?

.* (Greedy Match):

.* is a greedy quantifier, which means it tries to match as much as possible while still allowing the rest of the pattern to match.
It matches the longest possible sequence of characters that satisfies the overall pattern.

.*? (Lazy or Non-Greedy Match):

.*? is a non-greedy (lazy) quantifier, which means it tries to match as little as possible while still allowing the rest of the pattern to match.
It matches the shortest possible sequence of characters that satisfies the overall pattern.
```

```python
17. What does the . character normally match? What does it match if re.DOTALL is passed as 2nd
argument in re.compile()?


In regular expressions, the . (dot) character normally matches any character except a newline character (\n). It's a wildcard that represents a single character of any kind.

However, when you use the re.DOTALL flag as the second argument in re.compile(), it changes the behavior of the . (dot) character. The re.DOTALL flag is also sometimes referred to as re.S. When you enable this flag, the . character matches any character, including newline characters (\n).

Here's the difference in behavior:

Without re.DOTALL (default behavior):

. matches any character except a newline (\n) character.
Example:

python
Copy code
import re

text = "Hello\nWorld"
pattern = r".+"

result = re.findall(pattern, text)
print(result)
# Output: ['Hello']
In this example, without re.DOTALL, the . character matches everything up to the newline character, but it doesn't match the newline itself.

With re.DOTALL (or re.S) flag:

. matches any character, including newline (\n) characters.
Example:

python
Copy code
import re

text = "Hello\nWorld"
pattern = r".+"

# Using re.DOTALL flag
result = re.findall(pattern, text, re.DOTALL)
print(result)
# Output: ['Hello\nWorld']
In this example, with re.DOTALL, the . character matches everything, including newline characters, so it captures the entire text "Hello\nWorld."

Using re.DOTALL can be useful when you want to match patterns that span multiple lines or when you want to capture text that includes newline character
```

```python
19. What does passing re.VERBOSE as the 2nd argument to re.compile() allow to do?


Passing re.VERBOSE as the second argument to re.compile() in Python allows you to create regular expressions with improved readability and maintainability by enabling "verbose" mode. In verbose mode, you can add comments and whitespace to your regular expression patterns without affecting their functionality. This makes it easier to document and organize complex regular expressions.

Here's what using re.VERBOSE allows you to do:

Ignore Whitespace: You can use spaces, tabs, and line breaks to format your regular expression for better readability. In a non-verbose regular expression, these whitespace characters would be treated as literal characters, but in verbose mode, they are ignored.

Add Comments: You can add comments to explain parts of your regular expression using the # character. These comments are ignored by the regular expression engine.

Here's an example to illustrate how re.VERBOSE works:

python
Copy code
import re

# Non-verbose regular expression (hard to read)
pattern_non_verbose = r'\d{3}-\d{2}-\d{4}'

# Verbose regular expression (easier to read and comment)
pattern_verbose = r'''
    \d{3}    # Match 3 digits
    -        # Match a hyphen
    \d{2}    # Match 2 digits
    -        # Match another hyphen
    \d{4}    # Match 4 digits
'''

text = "123-45-6789"

# Compile both patterns and use them to match the text
regex_non_verbose = re.compile(pattern_non_verbose)
regex_verbose = re.compile(pattern_verbose, re.VERBOSE)

match_non_verbose = regex_non_verbose.match(text)
match_verbose = regex_verbose.match(text)

print(match_non_verbose.group())  # Output: 123-45-6789
print(match_verbose.group())      # Output: 123-45-6789
In this example, re.VERBOSE allows you to format and comment the regular expression pattern for better readability while still producing the same match result as the non-verbose pattern. This can be especially helpful when dealing with complex regular expressions.

```

```python
20. How would you write a regex that match a number with comma for every three digits? It must
match the given following:
    
'42'
'1,234'
'6,368,745'



import re

pattern = r'\d{1,3}(,\d{3})*'
text1 = '42'
text2 = '1,234'
text3 = '6,368,745'

match1 = re.fullmatch(pattern, text1)
match2 = re.fullmatch(pattern, text2)
match3 = re.fullmatch(pattern, text3)

print(match1)  # <re.Match object; span=(0, 2), match='42'>
print(match2)  # <re.Match object; span=(0, 5), match='1,234'>
print(match3)  # <re.Match object; span=(0, 9), match='6,368,745'>
```
