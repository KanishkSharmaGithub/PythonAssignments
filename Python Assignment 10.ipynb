```python
1. How do you distinguish between shutil.copy() and shutil.copytree()?


shutil.copy() Function:

shutil.copy() is used to copy a single file from one location to another.
It takes two arguments: the source file path and the destination file path.
The source file is copied to the destination file. If the destination file already exists, it will be overwritten.
It does not copy directories or their contents, only individual files.

import shutil

source_file = 'source.txt'
destination_file = 'destination.txt'

shutil.copy(source_file, destination_file)



shutil.copytree() Function:

shutil.copytree() is used to copy an entire directory, including its contents, to a new location.
It takes two arguments: the source directory path and the destination directory path.
The source directory and all of its contents are copied to the destination directory.
If the destination directory already exists, it will raise an error. You should ensure the destination directory doesn't exist
or remove it before using copytree().

import shutil

source_directory = 'source_dir'
destination_directory = 'destination_dir'

shutil.copytree(source_directory, destination_directory)
```

```python
3. What is the difference between the delete functions in the send2trash and shutil modules?


send2trash Module:

The send2trash module is designed to send files and directories to the operating system's trash or recycle bin (i.e., the system's native "delete" mechanism).
It provides a function called send2trash() that allows you to safely delete files and directories, which can be restored from the trash/recycle bin if needed.
It is platform-independent and works on various operating systems, including Windows, macOS, and Linux.

shutil Module:

The shutil module is a more general-purpose module for file operations, including copying, moving, and deleting files and directories.
It provides a function called shutil.rmtree() to delete directories and their contents recursively. It also provides os.remove() to delete individual files.
Unlike send2trash, shutil directly deletes files and directories without moving them to the trash, which means they cannot be easily restored.
```

```python
5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.



import os
import shutil

def search_and_copy_files(source_folder, destination_folder, file_extension):
    try:
        # Create the destination folder if it doesn't exist
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)

        # Walk through the source folder tree
        for root, _, files in os.walk(source_folder):
            for file in files:
                if file.endswith(file_extension):
                    source_path = os.path.join(root, file)
                    destination_path = os.path.join(destination_folder, file)

                    # Copy the file to the destination folder
                    shutil.copy2(source_path, destination_path)

        print(f"Files with the '{file_extension}' extension copied to {destination_folder}")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    source_folder = "source_folder_path"       # Replace with your source folder path
    destination_folder = "destination_folder"  # Replace with your destination folder path
    file_extension = ".pdf"                   # Replace with the desired file extension

    search_and_copy_files(source_folder, destination_folder, file_extension)
```
