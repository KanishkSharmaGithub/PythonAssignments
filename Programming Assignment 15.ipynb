```python
Please write a program using generator to print the numbers which can be divisible by 5 and
7 between 0 and n in comma separated form while n is input by console.


def divisible_5_7_generator(n):
    for i in range(n + 1):
        if i % 5 == 0 and i % 7 == 0:
            yield i

# Input from the user
n = int(input("Enter a value for 'n': "))

# Using the generator to generate the divisible numbers
result = divisible_5_7_generator(n)

# Printing the divisible numbers in comma-separated form
print("Numbers divisible by 5 and 7 between 0 and", n, "are:")
print(','.join(map(str, result)))
```

```python
Please write a program using generator to print the even numbers between 0 and n in comma
separated form while n is input by console.



def even_number_generator(n):
    for i in range(n + 1):
        if i % 2 == 0:
            yield i

# Input from the user
n = int(input("Enter a value for 'n': "))

# Using the generator to generate the even numbers
result = even_number_generator(n)

# Printing the even numbers in comma-separated form
print("Even numbers between 0 and", n, "are:")
print(','.join(map(str, result)))
```

```python
The Fibonacci Sequence is computed based on the following formula:
f(n)=0 if n=0
f(n)=1 if n=1
f(n)=f(n-1)+f(n-2) if n&gt;1
Please write a program using list comprehension to print the Fibonacci Sequence in comma
separated form with a given n input by console.



# Function to generate Fibonacci sequence using list comprehension
def generate_fibonacci_sequence(n):
    fib_sequence = [0 if i == 0 else 1 if i == 1 else fib_sequence[i-1] + fib_sequence[i-2] for i in range(n)]
    return fib_sequence

# Input from the user
n = int(input("Enter a value for 'n': "))

# Generating the Fibonacci sequence using list comprehension
fib_result = generate_fibonacci_sequence(n)

# Printing the Fibonacci sequence in comma-separated form
print("Fibonacci Sequence up to", n, "terms:")
print(','.join(map(str, fib_result)))
```

```python
Assuming that we have some email addresses in the &quot;username@companyname.com&quot; format,
please write program to print the user name of a given email address. Both user names and
company names are composed of letters only.


def extract_username(email):
    # Splitting the email address at the '@' symbol
    parts = email.split('@')

    # Extracting the username
    username = parts[0]

    return username

# Example usage:
email_address = "john@example.com"  # Replace this with any email address
username = extract_username(email_address)
print("Username:", username)
```

```python
Define a class named Shape and its subclass Square. The Square class has an init function
which takes a length as argument. Both classes have a area function which can print the area
of the shape where Shape&#39;s area is 0 by default.


class Shape:
    def __init__(self):
        pass

    def area(self):
        return 0

class Square(Shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length * self.length

# Create an instance of Square
square = Square(5)  # Pass the length of the square

# Calculate and print the area of the square
print("Area of the square:", square.area())

# Create an instance of Shape
shape = Shape()

# Calculate and print the area of the shape
print("Area of the shape:", shape.area())
```
