```python
1. Write a Python program to find sum of elements in list?


# Function to find the sum of elements in a list
def find_sum(lst):
    total = 0
    for element in lst:
        total += element
    return total

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Calculate and print the sum of the list elements
    total_sum = find_sum(numbers)
    print(f"The sum of the elements in the list is: {total_sum}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
2. Write a Python program to Multiply all numbers in the list?



# Function to multiply all numbers in a list
def multiply_list(lst):
    result = 1
    for number in lst:
        result *= number
    return result

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Calculate and print the product of the list elements
    product = multiply_list(numbers)
    print(f"The product of the numbers in the list is: {product}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
3. Write a Python program to find smallest number in a list?


# Function to find the smallest number in a list
def find_smallest_number(lst):
    if not lst:
        return "The list is empty."

    smallest = lst[0]
    for number in lst:
        if number < smallest:
            smallest = number
    return smallest

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Find and print the smallest number in the list
    smallest = find_smallest_number(numbers)
    print(f"The smallest number in the list is: {smallest}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
4. Write a Python program to find largest number in a list?


# Function to find the largest number in a list
def find_largest_number(lst):
    if not lst:
        return "The list is empty."

    largest = lst[0]
    for number in lst:
        if number > largest:
            largest = number
    return largest

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Find and print the largest number in the list
    largest = find_largest_number(numbers)
    print(f"The largest number in the list is: {largest}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
5. Write a Python program to find second largest number in a list?



# Function to find the second largest number in a list
def find_second_largest_number(lst):
    if len(lst) < 2:
        return "The list does not contain at least two numbers."

    largest = second_largest = float('-inf')

    for number in lst:
        if number > largest:
            second_largest = largest
            largest = number
        elif number > second_largest and number != largest:
            second_largest = number

    if second_largest == float('-inf'):
        return "There is no second largest number in the list."

    return second_largest

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Find and print the second largest number in the list
    second_largest = find_second_largest_number(numbers)
    print(f"The second largest number in the list is: {second_largest}")
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
6. Write a Python program to find N largest elements from a list?


# Function to find the N largest elements from a list
def find_n_largest_elements(lst, n):
    if n <= 0:
        return "N should be a positive integer."
    
    if n >= len(lst):
        return sorted(lst, reverse=True)

    return sorted(lst, reverse=True)[:n]

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Input N, the number of largest elements to find
    n = int(input("Enter the number of largest elements to find: "))

    # Find and print the N largest elements in the list
    largest_elements = find_n_largest_elements(numbers, n)
    print(f"The {n} largest elements in the list are: {largest_elements}")
except ValueError:
    print("Invalid input. Please enter a list of numbers and a valid positive integer for N.")
```

```python
7. Write a Python program to print even numbers in a list?


# Function to print even numbers in a list
def print_even_numbers(lst):
    print("Even numbers in the list:")
    for number in lst:
        if number % 2 == 0:
            print(number)

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Call the function to print even numbers
    print_even_numbers(numbers)
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
8. Write a Python program to print odd numbers in a List?


# Function to print odd numbers in a list
def print_odd_numbers(lst):
    print("Odd numbers in the list:")
    for number in lst:
        if number % 2 != 0:
            print(number)

# Input from the user
try:
    # Input space-separated numbers and convert them to a list of integers
    input_str = input("Enter a list of numbers separated by spaces: ")
    numbers = list(map(int, input_str.split()))

    # Call the function to print odd numbers
    print_odd_numbers(numbers)
except ValueError:
    print("Invalid input. Please enter a list of numbers separated by spaces.")
```

```python
9. Write a Python program to Remove empty List from List?



# Function to remove empty lists from a list of lists
def remove_empty_lists(lst):
    return [sublist for sublist in lst if sublist]

# Input from the user
try:
    # Input lists of lists, separated by spaces and convert them to a list of lists
    input_str = input("Enter a list of lists separated by spaces: ")
    list_of_lists = [eval(sublist) for sublist in input_str.split()]

    # Call the function to remove empty lists
    filtered_list = remove_empty_lists(list_of_lists)
    print("List after removing empty lists:", filtered_list)
except (ValueError, SyntaxError):
    print("Invalid input. Please enter a list of lists separated by spaces.")
```
