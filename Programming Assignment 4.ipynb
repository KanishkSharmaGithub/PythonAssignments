```python
1. Write a Python Program to Find the Factorial of a Number?


# Function to calculate the factorial of a number
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Input from the user
num = int(input("Enter a number: "))

# Calculate and print the factorial
fact = factorial(num)
print(f"The factorial of {num} is {fact}")
```

```python
2. Write a Python Program to Display the multiplication Table?



# Function to display the multiplication table
def multiplication_table(number, upto):
    print(f"Multiplication table for {number} up to {upto}:\n")
    for i in range(1, upto + 1):
        result = number * i
        print(f"{number} x {i} = {result}")

# Input from the user
num = int(input("Enter a number: "))
range_limit = int(input("Enter the range (up to which number you want to display the table): "))

# Display the multiplication table
multiplication_table(num, range_limit)
```

```python
3. Write a Python Program to Print the Fibonacci sequence?


# Function to generate and print the Fibonacci sequence
def fibonacci_sequence(n):
    fib_seq = []
    a, b = 0, 1

    while len(fib_seq) < n:
        fib_seq.append(a)
        a, b = b, a + b

    return fib_seq

# Input from the user
terms = int(input("Enter the number of terms in the Fibonacci sequence: "))

# Generate and print the Fibonacci sequence
fibonacci_seq = fibonacci_sequence(terms)
print("Fibonacci sequence:")
for term in fibonacci_seq:
    print(term, end=" ")
```

```python
4. Write a Python Program to Check Armstrong Number?


# Function to check if a number is an Armstrong number
def is_armstrong(number):
    # Calculate the number of digits in the given number
    num = number
    num_of_digits = len(str(num))
    
    # Calculate the sum of digits, each raised to the power of num_of_digits
    sum_of_digits = 0
    while num > 0:
        digit = num % 10
        sum_of_digits += digit ** num_of_digits
        num //= 10
    
    # Check if the number is an Armstrong number
    if number == sum_of_digits:
        return True
    else:
        return False

# Input from the user
num = int(input("Enter a number: "))

# Check if the number is an Armstrong number and print the result
if is_armstrong(num):
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```

```python
5. Write a Python Program to Find Armstrong Number in an Interval?


# Function to check if a number is an Armstrong number
def is_armstrong(number):
    num = number
    num_of_digits = len(str(num))
    sum_of_digits = 0

    while num > 0:
        digit = num % 10
        sum_of_digits += digit ** num_of_digits
        num //= 10

    return number == sum_of_digits

# Input from the user
lower_limit = int(input("Enter the lower limit of the interval: "))
upper_limit = int(input("Enter the upper limit of the interval: "))

# Find and print Armstrong numbers within the interval
print(f"Armstrong numbers between {lower_limit} and {upper_limit}:")
for num in range(lower_limit, upper_limit + 1):
    if is_armstrong(num):
        print(num)
```

```python
6. Write a Python Program to Find the Sum of Natural Numbers?


# Function to calculate the sum of natural numbers up to n
def sum_of_natural_numbers(n):
    if n < 1:
        return "Please enter a positive integer."

    # Using the formula for the sum of the first n natural numbers
    sum_n = n * (n + 1) // 2
    return sum_n

# Input from the user
num = int(input("Enter a positive integer: "))

# Calculate and print the sum of natural numbers up to num
result = sum_of_natural_numbers(num)
if isinstance(result, str):
    print(result)
else:
    print(f"The sum of natural numbers up to {num} is {result}.")
```
