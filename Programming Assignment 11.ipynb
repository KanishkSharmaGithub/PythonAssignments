```python
1. Write a Python program to find words which are greater than given length k?


# Function to find words greater than length k in a list
def find_words_greater_than_k(word_list, k):
    result = [word for word in word_list if len(word) > k]
    return result

# Input from the user
try:
    # Input a list of words separated by spaces
    input_str = input("Enter a list of words separated by spaces: ")
    word_list = input_str.split()

    # Input the minimum length k
    k = int(input("Enter the minimum length (k): "))

    # Find and print words greater than length k
    long_words = find_words_greater_than_k(word_list, k)
    
    if long_words:
        print(f"Words greater than length {k}: {long_words}")
    else:
        print(f"No words found greater than length {k}.")
except ValueError:
    print("Invalid input. Please enter a list of words and a valid integer for k.")
```

```python
2. Write a Python program for removing i-th character from a string?



# Function to remove the i-th character from a string
def remove_character_at_index(input_string, index):
    if 0 <= index < len(input_string):
        result = input_string[:index] + input_string[index + 1:]
        return result
    else:
        return "Index out of range."

# Input from the user
try:
    # Input a string
    input_string = input("Enter a string: ")
    
    # Input the index to remove
    index = int(input("Enter the index to remove (0-based): "))
    
    # Remove the character at the specified index
    modified_string = remove_character_at_index(input_string, index)
    
    if modified_string != "Index out of range.":
        print(f"String after removing character at index {index}: {modified_string}")
    else:
        print(modified_string)
except ValueError:
    print("Invalid input. Please enter a valid string and a valid integer index.")
```

```python
3. Write a Python program to split and join a string?


# Function to split a string and join it back
def split_and_join(input_string):
    # Split the input string into a list of words
    words = input_string.split()
    
    # Join the list of words back into a single string
    joined_string = ' '.join(words)
    
    return words, joined_string

# Input from the user
input_string = input("Enter a string: ")

# Call the function to split and join the string
split_words, joined_string = split_and_join(input_string)

print("Original String:", input_string)
print("Split Words:", split_words)
print("Joined String:", joined_string)
```

```python
4. Write a Python to check if a given string is binary string or not?


# Function to check if a string is a binary string
def is_binary_string(input_string):
    for char in input_string:
        if char not in '01':
            return False
    return True

# Input from the user
input_string = input("Enter a string: ")

# Check if the string is a binary string
if is_binary_string(input_string):
    print("The entered string is a binary string.")
else:
    print("The entered string is not a binary string.")
```

```python
5. Write a Python program to find uncommon words from two Strings?



# Function to find uncommon words from two strings
def find_uncommon_words(string1, string2):
    words1 = set(string1.split())
    words2 = set(string2.split())
    
    uncommon_words = words1.symmetric_difference(words2)
    return uncommon_words

# Input from the user
string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")

# Find and print uncommon words
uncommon_words = find_uncommon_words(string1, string2)

if uncommon_words:
    print("Uncommon words in the two strings:")
    for word in uncommon_words:
        print(word)
else:
    print("There are no uncommon words between the two strings.")
```

```python
6. Write a Python to find all duplicate characters in string?


# Function to find all duplicate characters in a string
def find_duplicate_characters(input_string):
    char_count = {}
    duplicates = []

    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char, count in char_count.items():
        if count > 1:
            duplicates.append(char)

    return duplicates

# Input from the user
input_string = input("Enter a string: ")

# Find and print duplicate characters
duplicate_chars = find_duplicate_characters(input_string)

if duplicate_chars:
    print("Duplicate characters in the string:", duplicate_chars)
else:
    print("No duplicate characters found in the string.")
```

```python
7. Write a Python Program to check if a string contains any special character?



import string

# Function to check if a string contains any special characters
def contains_special_characters(input_string):
    special_characters = set(string.punctuation)  # Set of special characters
    for char in input_string:
        if char in special_characters:
            return True
    return False

# Input from the user
input_string = input("Enter a string: ")

# Check if the string contains any special characters
if contains_special_characters(input_string):
    print("The string contains special characters.")
else:
    print("The string does not contain any special characters.")
```
