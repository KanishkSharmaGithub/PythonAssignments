```python
1. Add the current date to the text file today.txt as a string.


from datetime import date

# Get the current date as a string
current_date = date.today().strftime("%Y-%m-%d")

# Open 'today.txt' in append mode and write the current date
with open('today.txt', 'a') as file:
    file.write(current_date + '\n')

print("Current date has been added to 'today.txt'.")
```

```python
2. Read the text file today.txt into the string today_string


# Open 'today.txt' in read mode and read its contents into today_string
with open('today.txt', 'r') as file:
    today_string = file.read()

# Print the contents of today_string
print(today_string)

We open the file 'today.txt' in read mode ('r') using a with statement.

We read the contents of the file into the string today_string using the read() method.

We print the contents of today_string to the console.

After running this code, today_string will contain the contents of 'today.txt'.
```

```python
3. Parse the date from today_string.


from datetime import datetime

# Parse the date from today_string
parsed_date = datetime.strptime(today_string.strip(), "%Y-%m-%d")

# Print the parsed date
print("Parsed date:", parsed_date)

# Print the year, month, and day separately
print("Year:", parsed_date.year)
print("Month:", parsed_date.month)
print("Day:", parsed_date.day)

We import the datetime class from the datetime module.

We use datetime.strptime(today_string.strip(), "%Y-%m-%d") to parse the date from the today_string. The strip() method is used to remove any leading or trailing whitespace from the string, and "%Y-%m-%d" specifies the expected date format (in "YYYY-MM-DD" format).

We print the parsed date using print("Parsed date:", parsed_date).

We also print the year, month, and day separately using parsed_date.year, parsed_date.month, and parsed_date.day.

After running this code, you will have successfully parsed the date from today_string and can access its components as needed.
```

```python
4. List the files in your current directory.


import os

# Get the list of files in the current directory
current_directory = os.getcwd()
files = os.listdir(current_directory)

# Print the list of files
print("Files in the current directory:")
for file in files:
    print(file)
    
We import the os module to work with the file system.

We use os.getcwd() to get the current working directory.

We use os.listdir(current_directory) to obtain a list of files and directories in the current directory.

We print the list of files using a for loop.

Running this code will display the names of the files in your current directory.
    
```

```python
5. Create a list of all of the files in your parent directory (minimum five files should be available).



import os

# Get the parent directory
parent_directory = os.path.abspath(os.path.join(os.getcwd(), ".."))

# Get the list of files in the parent directory
files_in_parent = [f for f in os.listdir(parent_directory) if os.path.isfile(os.path.join(parent_directory, f))]

# Print the list of files in the parent directory
print("Files in the parent directory:")
for file in files_in_parent:
    print(file)
    
We import the os module to work with the file system.

We use os.getcwd() to get the current working directory, and then we use os.path.abspath(os.path.join(current_directory, "..")) to obtain the absolute path of the parent directory.

We use a list comprehension to iterate through the items in the parent directory and filter out only the files using os.path.isfile().

We print the list of files in the parent directory using a for loop.

Running this code will create a list of all the files in your parent directory and display their names.
```

```python
6. Use multiprocessing to create three separate processes. Make each one wait a random number of
seconds between one and five, print the current time, and then exit.


import multiprocessing
import random
import time
from datetime import datetime

def worker(process_id):
    sleep_time = random.randint(1, 5)
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    print(f"Process {process_id}: Waiting for {sleep_time} seconds at {current_time}")
    time.sleep(sleep_time)
    print(f"Process {process_id}: Done at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

if __name__ == "__main__":
    processes = []

    # Create three separate processes
    for i in range(3):
        process = multiprocessing.Process(target=worker, args=(i,))
        processes.append(process)
        process.start()

    # Wait for all processes to finish
    for process in processes:
        process.join()

    print("All processes have finished.")
    
```

```python
8. What day of the week was your day of birth?


from datetime import date

# Replace these values with your birthdate (year, month, day)
year_of_birth = 2000
month_of_birth = 1
day_of_birth = 1

# Create a date object for your date of birth
birth_date = date(year_of_birth, month_of_birth, day_of_birth)

# Get the day of the week (0 = Monday, 1 = Tuesday, ..., 6 = Sunday)
day_of_week = birth_date.weekday()

# Define a list of weekday names
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Print the day of the week for your birthdate
print(f"Your date of birth ({birth_date}) falls on a {weekday_names[day_of_week]}.")
```

```python
9. When will you be (or when were you) 10,000 days old?


from datetime import date, timedelta

# Replace these values with your birthdate (year, month, day)
year_of_birth = 2000
month_of_birth = 1
day_of_birth = 1

# Define the number of days you want to add (10,000 days)
days_to_add = 10000

# Create a date object for your date of birth
birth_date = date(year_of_birth, month_of_birth, day_of_birth)

# Calculate the date when you will be (or were) 10,000 days old
target_date = birth_date + timedelta(days=days_to_add)

# Print the target date
print(f"You will be (or were) 10,000 days old on {target_date}.")
```
