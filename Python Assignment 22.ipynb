```python
1. What is the result of the code, and explain?

&gt;&gt;&gt; X = &#39;iNeuron&#39;
&gt;&gt;&gt; def func():
print(X)

&gt;&gt;&gt; func()

You assign the string 'iNeuron' to the variable X in the global scope.

You define a function named func().

Inside the func() function, you attempt to print the value of the variable X.

When you call func() using func(), it will attempt to print the value of X. However, there is no X defined within the local scope of the func() function.

Python uses a mechanism called "variable scoping." In this case, since X is not defined as a local variable within the func() 
function, Python will look for it in the nearest enclosing scope, which is the global scope. It finds X in the global scope and 
prints its value, which is 'iNeuron'

Result will be:
    
    ineuron
```
