```python
1. Write a Python program to Extract Unique values dictionary values?


# Function to extract unique values from dictionary values
def extract_unique_values(dictionary):
    unique_values = set()
    for value in dictionary.values():
        if isinstance(value, list):
            unique_values.update(value)
        else:
            unique_values.add(value)
    return list(unique_values)

# Input from the user (a sample dictionary)
sample_dict = {
    'a': [1, 2, 3],
    'b': [2, 3, 4],
    'c': 5,
    'd': 4
}

# Extract and print unique values from the dictionary
unique_values = extract_unique_values(sample_dict)
print("Unique values in the dictionary:", unique_values)
```

```python
2. Write a Python program to find the sum of all items in a dictionary?



# Function to find the sum of all items in a dictionary
def sum_of_dict_values(dictionary):
    total_sum = sum(dictionary.values())
    return total_sum

# Input from the user (a sample dictionary)
sample_dict = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

# Find and print the sum of all items in the dictionary
total_sum = sum_of_dict_values(sample_dict)
print("Sum of all items in the dictionary:", total_sum)
```

```python
3. Write a Python program to Merging two Dictionaries?


# Function to merge two dictionaries
def merge_dicts(dict1, dict2):
    merged_dict = {**dict1, **dict2}
    return merged_dict

# Input from the user (two sample dictionaries)
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

# Merge the two dictionaries and print the result
merged_dict = merge_dicts(dict1, dict2)
print("Merged Dictionary:", merged_dict)
```

```python
4. Write a Python program to convert key-values list to flat dictionary?



# Function to convert a list of key-value pairs to a flat dictionary
def list_to_flat_dict(kv_list):
    flat_dict = {}
    for pair in kv_list:
        if len(pair) == 2:
            key, value = pair
            flat_dict[key] = value
    return flat_dict

# Input from the user (a sample list of key-value pairs)
sample_kv_list = [("a", 1), ("b", 2), ("c", 3)]

# Convert the list to a flat dictionary and print the result
flat_dict = list_to_flat_dict(sample_kv_list)
print("Flat Dictionary:", flat_dict)
```

```python
5. Write a Python program to insertion at the beginning in OrderedDict?


from collections import OrderedDict

# Function to insert a key-value pair at the beginning of an OrderedDict
def insert_at_beginning(od, key, value):
    od.move_to_end(key, last=False)
    od[key] = value

# Input from the user (a sample OrderedDict)
sample_ordered_dict = OrderedDict([
    ('b', 2),
    ('c', 3),
    ('d', 4)
])

# Input key and value to insert
key_to_insert = 'a'
value_to_insert = 1

# Insert the key-value pair at the beginning of the OrderedDict and print the result
insert_at_beginning(sample_ordered_dict, key_to_insert, value_to_insert)
print("OrderedDict after insertion at the beginning:", sample_ordered_dict)
```

```python
7. Write a Python program to sort Python Dictionaries by Key or Value?


# Function to sort a dictionary by keys
def sort_dict_by_key(input_dict):
    sorted_dict = dict(sorted(input_dict.items()))
    return sorted_dict

# Function to sort a dictionary by values
def sort_dict_by_value(input_dict):
    sorted_dict = dict(sorted(input_dict.items(), key=lambda item: item[1]))
    return sorted_dict

# Input from the user (a sample dictionary)
sample_dict = {
    'b': 2,
    'c': 1,
    'a': 3
}

# Sort the dictionary by keys and print the result
sorted_by_key = sort_dict_by_key(sample_dict)
print("Dictionary sorted by keys:", sorted_by_key)

# Sort the dictionary by values and print the result
sorted_by_value = sort_dict_by_value(sample_dict)
print("Dictionary sorted by values:", sorted_by_value)
```
